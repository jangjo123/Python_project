```python
import pygame
from pygame.locals import *
from enum import Enum, auto

class State(Enum):
    Idle = auto()
    Run = auto()
    Jump = auto()

_state = State.Idle

screen = pygame.display.set_mode((800,600))
pygame.init()

running = True

def IDLE():
    return

def RUN():
    return

def JUMP():
    return

def Get_Png(PngName):

    return pygame.image.load(PngName+".png")

def Animation(state):
    if state == State.Run:
        print("Move")

    elif _state == State.Jump:
        print("Jump")

    else: None#print("Idle")

def Input(eventKey):

    if event.type == KEYDOWN:
        if event.key == K_RIGHT:
            _state = State.Run
            RUN()
                
        elif event.key == K_LEFT:
            _state = State.Run
            RUN()

        elif event.key == K_SPACE:
            _state = State.Jump
            JUMP()

        else:
            _state = State.Idle
            IDLE()
        return _state

def Image_print(image, x, y):
    for i in range(0, len(image)):
        screen.blit(image[i], [ x, y ] )
    
    

Idle_anims = [Get_Png("Player_Idle1"), Get_Png("Player_Idle2")]
Idle_anim = []
Idle_anim.append(pygame.transform.scale(Idle_anims[0], (70, 60)))
Idle_anim.append(pygame.transform.scale(Idle_anims[1], (70, 60)))
# count = 0

while running:
    for event in pygame.event.get():
        _state = Input(event)
        Animation(_state)

        if event.type == pygame.QUIT:
            running = False

        for i in range(0,len(Idle_anim)):
             
            player = Idle_anim
            Test = Image_print(player, 200, 250)
            
        
        pygame.display.update()
    

pygame.quit()
```

```python
import pygame
import sys

SCREEN_WIDTH = 640
SCREEN_HEIGHT = 480

white = (255, 255, 255)
black = (0, 0, 0)

pygame.init()
pygame.display.set_caption("Simple PyGame Example")
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

pos_x = 200
pos_y = 200

clock = pygame.time.Clock()
while True:
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            sys.exit()

    key_event = pygame.key.get_pressed()
    if key_event[pygame.K_LEFT]:
        pos_x -= 1

    if key_event[pygame.K_RIGHT]:
        pos_x += 1

    if key_event[pygame.K_UP]:
        pos_y -= 1

    if key_event[pygame.K_DOWN]:
        pos_y += 1

    screen.fill(black)
    pygame.draw.circle(screen, white, (pos_x, pos_y), 20)
    pygame.display.update()
```
