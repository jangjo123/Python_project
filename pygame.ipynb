```python
import pygame
from pygame.locals import *
from enum import Enum, auto

class State(Enum):
    Idle = auto()
    Run = auto()
    Jump = auto()

_state = State.Idle

screen = pygame.display.set_mode((800,600))
pygame.init()

running = True

def Animation(state):
    if state == State.Run:
        print("Move")

    elif _state == State.Jump:
        print("Jump")

    else: None#print("Idle")

def Input(eventKey):

    if event.type == KEYDOWN:
        if event.key == K_RIGHT:
            _state = State.Run
                
        elif event.key == K_LEFT:
            _state = State.Run

        elif event.key == K_SPACE:
            _state = State.Jump

        else:
            _state = State.Idle
        return _state    

while running:
    for event in pygame.event.get():
        _state = Input(event)
        Animation(_state)

        if event.type == pygame.QUIT:
            running = False
        pygame.display.update()
    

pygame.quit()
```
